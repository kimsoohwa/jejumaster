```python

#지역별 충전기 년도 총합계 출력하기
import pandas as pd

PATH = './sample_data/'
FILE = '한국전력공사_지역별 전기차 충전소 현황정보_20230718.csv'

file = PATH+FILE
df = pd.read_csv(file, encoding = 'utf-8') #cp949: 윈도우OS 기본 인코딩 포맷
# df.head(3)#앞에서부터 3개만 보여라
df.head()

# '지역' 열을 기준으로 데이터프레임을 그룹화하고 2016년부터 2023년까지의 데이터를 합산
df['합계'] = df[['2016', '2017', '2018', '2019', '2020', '2021', '2023']].sum(axis=1)

# '지역'과 '합계' 열로 구성된 새로운 데이터프레임 생성
df_total_grouped = df[['지역', '합계']]


# '합계' 열을 기준으로 데이터프레임을 크기순으로 정렬
df_sorted = df_total_grouped.sort_values(by='합계', ascending=False)

# 결과를 출력
df_sorted
```

```python
df_sorted.columns
#df_total_grouped.columns
```

```python
#막대그래프로 시각화
import matplotlib.pyplot as plt

# 코랩에서 한글 폰트 설치한 후 사용하기!!!
# plt.rc('font', family='cmb10')                  # (코랩)한글 폰트
plt.rc('font', family='NanumBarunGothic')       # (코랩)한글 폰트
# plt.rc('font', family='Malgun Gothic')

# 1.그래프 데이터 지정하기(X축, Y축)
x = df_sorted['지역']    # 지역
y = df_sorted['합계']   # 합계

# 2.그래프 그리기
plt.bar(x, y)


# 3. 그래프 꾸미기
# plt.title('지역별 전기차 충전소 현황') # 그래프 제목
plt.title('지역별 전기차 충전소 현황') # 그래프 제목
plt.xlabel('지역')                    # x축 이름
plt.ylabel('충전기 수량')              # y축 이름
plt.xticks(rotation = 45)             # X축 레이블 회전
plt.show()                            # 그래프 출력하기
```

```python
#그래프 꾸미기
import matplotlib.pyplot as plt

# 1.그래프 데이터 지정하기(X축, Y축)
x = df_sorted['지역']    # 지역
y = df_sorted['합계']   # 합계

# 2.그래프 그리기
plt.figure(figsize = (13, 5), facecolor = 'palegreen') # 그래프 크기와 색상
plt.bar(x, y, color = 'blueviolet')  # 막대 그래프 그리기

# 3. 그래프 꾸미기
plt.title('"지역별 전기차 충전소 현황"', pad = 20,                # 그래프 제목
          fontdict = {'size':30, 'color':'black', 'weight':'bold'})
plt.xlabel('지역', fontdict = {'size': 20, 'color': 'purple', 'weight': 'bold'})
plt.ylabel('충전기 수(대)', fontdict = {'size': 20, 'color': 'purple', 'weight': 'bold'}, labelpad = 20)
plt.xticks(rotation = 30)                   # X축 레이블 회전
plt.grid(True, axis = 'y', alpha = 0.5)    # 그래프에 그리드 표현하기

plt.show() # 그래프 그리기
```

```python
#선그래프로 시각화
# 1.그래프 데이터 지정하기(X축, Y축)
x = df_sorted['지역']    # 지역
y = df_sorted['합계']   # 합계

# 2.그래프 그리기 (선그래프 그리기)
plt.figure(figsize = (15, 5), facecolor = 'lightsalmon') # 그래프 크기와 색상
plt.plot(x, y, color='crimson', marker='o', linestyle='solid',
         linewidth=2, markersize=8, markerfacecolor='red')

# 3. 그래프 꾸미기
plt.title('"지역별 전기차 충전소 현황""', pad = 20,                # 그래프 제목
          fontdict = {'size':30, 'color':'black', 'weight':'bold'})
plt.xlabel('지역', fontdict = {'size': 20, 'color': 'purple', 'weight': 'bold'})
plt.ylabel('충전기 수(대)', fontdict = {'size': 20, 'color': 'purple', 'weight': 'bold'}, labelpad = 20)
plt.xticks(rotation = 45)                   # X축 레이블 회전
plt.grid(True, axis = 'y', alpha = 0.5)    # 그래프에 그리드 표현하기

plt.show() # 그래프 그리기
```

```python
# 워드클라우드로 시각화하기
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# 1. (win)폰트 지정: 자신의 컴퓨터 환경에 맞는 한글 폰트 경로
# font_path = 'malgun'  # C:/Windows/Fonts/
# font_path = 'HMFMOLD' # C:/Windows/Fonts/HMKMMAG.TTF     #HMFMOLD.TTF

# 1. (코랩)폰트 지정: 자신의 컴퓨터 환경에 맞는 한글 폰트 경로
font_path = '/usr/share/fonts/truetype/nanum/NanumGothicBold.ttf'

# 2.전처리된 데이터 지정하기
names = df_sorted['지역']     # 지역
counts = df_sorted['합계']  # 인구수
data = dict(zip(names, counts)) # 워드클라우드는 딕셔너리 형태로 만들기

# 3.워드클라우드 그래프로 시각화하기
wc = WordCloud(width = 1000, height = 600,
               background_color="beige", font_path=font_path)#바탕색을 바꿀 수 있음
plt.imshow(wc.generate_from_frequencies(data))
plt.axis("off")#X,Y축 안나오게 지정
plt.show()

# 4.파일로 저장하기
wc.to_file('./전기차충전기_워드클라우드.png')#이미지가 저장됨

```
